```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

plt.style.use('dark_background')

%matplotlib inline
```

```python
df = pd.read_csv('./data/loan_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
df[df['credit.policy']==1]['fico'].hist(alpha=0.5,color='white', bins=30,label='Credit.Policy=1')
df[df['credit.policy']==0]['fico'].hist(alpha=0.5,color='red', bins=30,label='Credit.Policy=0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(10,6))
df[df['not.fully.paid']==1]['fico'].hist(alpha=0.5,color='white', bins=30,label='not.fully.paid=1')
df[df['not.fully.paid']==0]['fico'].hist(alpha=0.5,color='red', bins=30,label='not.fully.paid=0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='purpose', hue='not.fully.paid')
```

```python
sns.jointplot(data=df, x='fico', y='int.rate')
```

```python
plt.figure(figsize=(12, 8))
sns.lmplot(y='int.rate',x='fico',data=df,hue='credit.policy', col='not.fully.paid')
```

```python
train_data = pd.get_dummies(df, columns=['purpose'], drop_first=True).astype('int')
train_data.info()
```

```python
X = train_data.drop('not.fully.paid', axis=1)
Y = train_data['not.fully.paid']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
predictions = dtree.predict(x_test)

print(f"Classification Report:\n{classification_report(y_test, predictions)}\n")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, predictions)}")
```

```python
rfc = RandomForestClassifier(n_estimators=500)
rfc.fit(x_train, y_train)
predictions = rfc.predict(x_test)

print(f"Classification Report:\n{classification_report(y_test, predictions)}\n")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, predictions)}")
```
